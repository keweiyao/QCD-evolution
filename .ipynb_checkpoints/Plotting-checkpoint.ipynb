```python
def HQET(z, r):
    return r*z*(1-z)**2/(1-(1-r)*z)**6 * (
        6. 
      - 18.*(1.-2*r)*z 
      + (21.-74*r+68*r**2)*z**2
      - 2*(1.-r)*(6-19*r+18*r**2)*z**3
      + 3.*(1.-r)**2 * (1-2*r+2*r**2)*z**4
    )
```

```python
az = 0.118
Mz = 91.7
Mb = 4.5
Mc = 1.3
b5 = 4*np.pi/(11-2/3.*5)
b4 = 4*np.pi/(11-2/3.*4)
b3 = 4*np.pi/(11-2/3.*3)
Lambda2_5 = Mz**2/np.exp(b5/az)
a5_Mb = b5/np.log((2*Mb)**2/Lambda2_5)
Lambda2_4 = (2*Mb)**2/np.exp(b4/a5_Mb)
a4_Mc = b4/np.log((2*Mc)**2/Lambda2_4)
Lambda2_3 = (2*Mc)**2/np.exp(b3/a4_Mc)
print(Lambda2_3**.5, Lambda2_4**.5, Lambda2_5**.5)
@np.vectorize
def alphas(Q2):
    if (Q2<(2*Mc)**2):
        return b3/np.log(Q2/Lambda2_3)
    elif (Q2<(2*Mb)**2):
        return b4/np.log(Q2/Lambda2_4)
    else:
        return b5/np.log(Q2/Lambda2_5)
    
@np.vectorize
def alphas2(Q2):
    return b3/np.log(Q2/.2**2)
```

```python
Q2 = np.exp(np.linspace(np.log(.32), np.log(1e3),100))**2
plt.plot(Q2, alphas(Q2))
plt.plot(Q2, alphas2(Q2), 'k--')
plt.semilogx()
```
